# Estatística Aplicada

**Efetividade de Duas Estratégias de Ensino**

Imagine que uma escola esteja avaliando a eficácia de duas estratégias de ensino de matemática para alunos do ensino médio. Eles querem determinar se há uma diferença significativa no desempenho médio dos alunos entre as duas estratégias.

# Hipóteses:

* Hipótese nula (H0): A média das notas dos alunos na estratégia A é igual à média das notas dos alunos na estratégia B.
* Hipótese alternativa (H1): A média das notas na Estratégia B é maior do que a média das notas na Estratégia A.

# **Dados:**

* Amostra da Estratégia A: Notas de 50 alunos que receberam a Estratégia A.
* Amostra da Estratégia B: Notas de 50 alunos que receberam a Estratégia B.

Usaremos um teste Z para comparar as médias das notas entre as duas amostras.

Se o p-valor do teste Z for menor que um nível de significância pré-determinado (por exemplo, α = 0.05), rejeitamos a hipótese nula e concluímos que há uma diferença significativa nas médias das notas entre as duas estratégias de ensino.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import norm
```

Os dados são criados a seguir:

```python
# Definindo médias e desvios padrão para as notas nas duas estratégias
media_estrategia_A = 70
desvio_padrao_estrategia_A = 10

media_estrategia_B = 75
desvio_padrao_estrategia_B = 12

# Gerando as amostras de notas para cada estratégia de ensino da nossa base
np.random.seed(0)  # Para garantir a reprodutibilidade dos resultados
amostra_estrategia_A = np.random.normal(loc=media_estrategia_A, scale=desvio_padrao_estrategia_A, size=50)
amostra_estrategia_B = np.random.normal(loc=media_estrategia_B, scale=desvio_padrao_estrategia_B, size=50)

print("Notas da Estratégia A:", amostra_estrategia_A[:5])
print("Notas da Estratégia B:", amostra_estrategia_B[:5])
```

# 1) De acordo com as informações analisadas o nosso teste é unilateral á direita, esquerda ou bicaudal?




Para este caso, estamos utilizando um teste unilateral à direita. 
Essa escolha se justifica diretamente pela nossa hipótese alternativa (H1). Como o objetivo é verificar se a Estratégia B apresenta resultados superiores à Estratégia A, estamos analisando se a diferença entre as médias é positiva >0.

Sempre que a hipótese alternativa busca identificar se um parâmetro é maior que o outro, a região crítica de rejeição se concentra na cauda superior da distribuição. Portanto, como o foco está exclusivamente no aumento de desempenho da Estratégia B, o teste é classificado como unilateral à direita.

# 2) Calculo das médias para as duas amostragens e as variâncias + insights.

```python
# médias das amostras
media_A = np.mean(amostra_estrategia_A)
media_B = np.mean(amostra_estrategia_B)

# variâncias das amostras
variancia_A = np.var(amostra_estrategia_A, ddof=1)
variancia_B = np.var(amostra_estrategia_B, ddof=1)

print("Média Estratégia A:", media_A)
print("Variância Estratégia A:", variancia_A)

print("Média Estratégia B:", media_B)
print("Variância Estratégia B:", variancia_B)
```

Principais insights do resultado:
Ao comparar os resultados de média e variância entre as duas abordagens, conseguimos extrair alguns insights valiosos:
Primeiro, notamos que a Estratégia B apresenta uma média superior (74,75 contra 71,41 da Estratégia A). Essa vantagem de aproximadamente 3,34 pontos sugere que os alunos do grupo B tiveram, em termos gerais, um desempenho melhor. Embora esse dado reforce a nossa hipótese alternativa, ele ainda atua apenas como um indicativo; precisamos da validação do teste Z para confirmar se essa diferença é estatisticamente significativa ou se foi fruto do acaso.

Outro ponto importante é a variabilidade. A Estratégia A registrou uma variância maior (129,27), o que aponta para notas mais dispersas e resultados mais heterogêneos entre os alunos. Já a Estratégia B mostrou-se mais consistente, com uma variância menor (110,47), indicando que o desempenho do grupo foi mais uniforme e concentrado próximo à média.

Em resumo: a Estratégia B não só entrega uma média mais alta, como também oferece maior previsibilidade. 

# 3) Resultados da estatística do teste Z, p value e se rejeitaremos ou não a hipótese nula + Justificativa.

```python
# médias
media_A = np.mean(amostra_estrategia_A)
media_B = np.mean(amostra_estrategia_B)

# variâncias
var_A = np.var(amostra_estrategia_A, ddof=1)
var_B = np.var(amostra_estrategia_B, ddof=1)

# tamanhos das amostras
n_A = len(amostra_estrategia_A)
n_B = len(amostra_estrategia_B)

# erro padrão da diferença das médias
erro_padrao = np.sqrt((var_A/n_A) + (var_B/n_B))

# estatística Z
z_stat = (media_B - media_A) / erro_padrao

# p-value (teste unilateral à direita)
p_value = 1 - stats.norm.cdf(z_stat)

print("Estatística Z:", z_stat)
print("p-value:", p_value)

# decisão do teste
alpha = 0.05

if p_value < alpha:
    print("Rejeitamos a hipótese nula (H0)")
else:
    print("Não rejeitamos a hipótese nula (H0)")
```

Após os cálculos, chegamos a uma estatística de teste Z de 1,5267, o que resulta em um p-value de 0,0634. 
Como estamos realizando um teste unilateral à direita com um nível de significância de 5%, a regra de decisão é clara: como o nosso p-value é superior ao nível de significância, não temos evidências estatísticas suficientes para rejeitar a hipótese nula.

Na prática, isso significa que, apesar de a Estratégia B ter apresentado uma média nominalmente maior na amostra, essa diferença de 3,34 pontos não é grande o suficiente para descartarmos a possibilidade de ser apenas fruto da variabilidade natural dos dados.

# 4) Gráfico da distribuição da estatística do teste Z e o que pode ser observado através dele.

```python
# valor da estatística Z obtida
z_stat = 1.5267455833477102

# eixo x da distribuição normal
x = np.linspace(-4, 4, 1000)

# função densidade da normal padrão
y = stats.norm.pdf(x)

plt.figure(figsize=(8,5))
plt.plot(x, y, label="Distribuição Normal Padrão (Z)")
plt.axvline(z_stat, linestyle="--", label=f"Z calculado = {z_stat:.2f}")

plt.title("Distribuição da Estatística do Teste Z")
plt.xlabel("Valores de Z")
plt.ylabel("Densidade de Probabilidade")
plt.legend()

plt.show()
```

O gráfico mostra a distribuição Z, onde a linha vertical marca o nosso Z calculado de 1,53. Note que ele não avança o suficiente para a extremidade da curva, permanecendo fora da região crítica.

Isso visualiza o que o p-valor de 0,063 já nos indicou: a diferença entre as estratégias ainda está dentro de uma margem comum de variação. Como o valor não atingiu a zona de rejeição, o gráfico confirma que não há evidência estatística para considerar a Estratégia B superior ao nível de 5% de significância.
